```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (right + left)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
for _ in range(5):
    arr = list(map(int, input().split()))
    target = int(input())
    print(binary_search(arr,0, len(arr) - 1, target))
    print("_________")
```

```python
def count(arr, x):
    ans = 0
    for item in arr:
        if item == x:
            ans += 1
    return ans
```

```python
def bubble_sort(arr):

    k = 1
    while True:
        swap = False
        if k == len(arr):
            return arr
        for i in range(len(arr) - k):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swap = True
        if not swap:
            return arr
        k += 1

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for it in range(min_idx, n):
            if arr[it] < arr[min_idx]:
                min_idx = it
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
#10
arr = list(map(int,input().split()))
k = int(input())

for i in range(1,len(arr)):
    arr[i] += arr[i - 1]
arr = [0] + arr
ans = 0
for i in range(len(arr)):
    for j in range(i + 1,len(arr)):
        if ans < (arr[j] - arr[i]) % k:
            ans = (arr[j] - arr[i]) % k

print(ans)
```
