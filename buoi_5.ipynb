```python
def is_prime(n):
    div = 2
    flag = True
    while div**2 < n:
        if n % div == 0:
            return False
        div += 1
    return flag
num = int(input())
print(is_prime(num))
```

```python
def is_per(n):
    div = []
    for i in range(1 , n):
        if n % i == 0:
            div.append(i)
    sum_fac = sum(div)
    if sum_fac == n:
        return True
    else:
        return False


num = int(input())
print(is_per(num))
```

```python
def find_num(lst, k):
    for idx in range(len(lst)):
        if lst[idx] == k:
            return idx + 1
    return -1

list_num = list(map(int, input().split()))
k = int(input())

print(find_num(list_num, k))
```

```python
def operator(a , b , oper):
    if oper == '+':
        return a + b
    elif oper == '-':
        return a - b
    elif oper == '*':
        return a * b
    elif oper == '/':
        return a / b

a , oper, b = input().split()

print(operator(int(a), int(b) ,oper))
```

```python
def bir(n):
    re_bir = ''
    while n > 0:
        if n % 2 == 0:
            re_bir += '0'
            n //= 2
        else:
            re_bir += '1'
            n //= 2
    return re_bir[:: -1]
a, b = map(int, input().split())

a_bir = bir(a)
b_bir = bir(b)

lent = abs(len(a_bir) - len(b_bir))
if len(a_bir) > len(b_bir):
    b_bir = '0' * lent + b_bir
else:
    a_bir = '0' * lent + a_bir

ans = 0
for i in range(len(a_bir)):
    if a_bir[i] != b_bir[i]:
        ans +=1
print(ans)

```

```python
def sum_digit(n):
    sum = 0
    for item in str(n):
        sum += int(item)
    return sum

print(sum_digit(23))
```

```python
def check(word_1, word_2):
    a = [0] * 26
    b = [0] * 26

    for item in word_1:
        a[ord(item) - 97] += 1
    for it in word_2:
        b[ord(it) - 97] += 1

    a = a.sort()
    b = b.sort()

    if a == b:
        return True
    else:
        return False
word_1 , word_2 = input().split()
print(check(word_1, word_2))

```
