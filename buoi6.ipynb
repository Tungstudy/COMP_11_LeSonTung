```python
#W6A4
arr = input()
my_list = list(arr.split(" "))

dic = dict()

for item in my_list:
    idx = item.find(":")
    key = item[:idx]
    val = item[idx + 1:]
    if key in dic:
        dic[key].append(val)
    else:
        dic[key] = [val]
print(dic)
```

```python
#W6A5
n = list(map(int, input().split()))

dic = {
    'positive' :0,
    'zeros': 0,
    'negatives': 0
}
for item in n:
    if item > 0:
        dic['positive'] += 1
    elif item == 0:
        dic['zeros'] += 1
    else:
        dic['negatives'] += 1
print(dic)
```

```python
#W6A10
arr = list(map(int, input().split()))
k = int(input())
len_arr = len(arr)
lst_ans = []
for idx in range(len_arr):
    if arr[idx] == k // 2:
        arr[idx] = 0

for i in range(len_arr):
    if (k - arr[i]) in arr[i:]:
        lst_ans.append((arr[i], k - arr[i]))
print(list(set(lst_ans)))
```

```python
#W6A11
arr = list(map(int, input().split()))
odd = list()
even = list()
for i in range(len(arr)):
    if arr[i] % 2 == 0:
        even.append(arr[i])
    else:
        odd.append(arr[i])
print(tuple(odd))
print(tuple(even))
```

```python
#W6A12
arr = list(map(int, input().split()))

count_arr = [0] *( max(arr) + 1)
for item in arr:
    count_arr[item] += 1
temp = 0
ans = 0
for i in range(len(count_arr)):
    if count_arr[i] > temp:
        temp = count_arr[i]
        ans = i

print(ans)
```

```python
#W6A14
arr_1 = list(map(int, input().split()))
arr_2 = list(map(int, input().split()))

lst_1 = list(set(arr_1))
lst_2 = list(set(arr_2))

dic = dict()
for i in lst_1:
    dic[i] = 1
ans = []
for item in lst_2:
    if item in dic:
        ans.append(item)
print(ans)
```
