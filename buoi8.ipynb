```python
#1
class Cylinder:
    pi = 3.14
    def __init__(self,radius, height):
        self.radius = radius
        self.height = height
    def area_suf(self):
        print( f'{((self.radius**2)*2*Cylinder.pi + 2*Cylinder.pi*self.radius*self.height):.2f}')
    def volume(self):
        ans = self.height * (self.radius**2)* Cylinder.pi
        print(f'{ans:.2f}')

r, height = map(int, input().split())
cylin = Cylinder(r, height)
cylin.area_suf()
cylin.volume()
```

```python
#7
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)
class Fraction:
    def __init__(self, ts, ms):
        if ms < 0:
            self.ts = -ts
            self.ms = -ms
        else:
            self.ts = ts
            self.ms = ms
    def __str__(self):
        return f'{self.ts}/{self.ms}'
    def __add__(self, other):
        tu_so = self.ts*other.ms + self.ms*other.ts
        mau_so = self.ms * other.ms
        div = gcd(tu_so , mau_so)
        return Fraction(tu_so // div, mau_so // div)
    def __sub__(self, other):
        tu_so = self.ts*other.ms - self.ms*other.ts
        mau_so = self.ms * other.ms
        div = gcd(tu_so , mau_so)
        return Fraction(tu_so // div, mau_so // div)
    def __mul__(self, other):
        tu_so = self.ts * other.ts
        mau_so = self.ms * other.ms
        div = gcd(tu_so, mau_so)
        return Fraction(tu_so//div, mau_so//div)
    def __truediv__(self, other):
        tu_so = self.ts * other.ms
        mau_so = self.ms * other.ts
        div = gcd(tu_so, mau_so)
        return Fraction(tu_so//div, mau_so//div)

k = Fraction(5,6)
j = Fraction(4,3)
print(k / j)
```

```python
#8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def __str__(self):
        return f'{self.owner} {self.balance}'
    def deposite(self, amt):
        self.balance += amt
        return self.balance
    def withdraw(self, amount):
        if self.balance < amount:
            return 'Invalid'
        else:
            self.balance -= amount
            return self.balance
```

```python
#10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def __str__(self):
        return f'{self.title} - {self.author} ({self.year})'

class Library:
    def __init__(self):
        self.books = []
    def add_book(self, book):
        self.books.append(book)

    def find_by_author(self, name):

        results = [b for b in self.books if b.author == name]
        for book in results:
            print(book)

lib = Library()

b1 = Book('Tung', 'Le', 2025)
b2 = Book('Son', 'Le', 1990)

lib.add_book(b1)
lib.add_book(b2)

print("--- Danh sách sách của tác giả 'Le': ---")
lib.find_by_author('Le')
```

```python
#12
class ComplexNum:
  def __init__(self, re, im):
    self.re = re
    self.im = im
  def __mul__(self, other):
    real = (self.re * other.re) - (self.im * other.im)
    imagin = (self.re*other.im) + (self.im * other.re)
    k = ComplexNum(real, imagin)
    return k
  def __str__(self):
    return f'{self.re} + {self.im}i'
```

```python
#13
import math
class Triangle:
    def __init__(self, a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def get_area(self):
        p = (self.a + self.b + self.c) / 2
        try:
            return math.sqrt((p-self.a)*(p-self.b)*(p-self.c)*p)
        except ValueError:
            return -1
```
